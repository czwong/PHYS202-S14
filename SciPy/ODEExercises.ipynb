(a) Use Euler's method to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, that is y(1), using a first a step size of 0.25, and then 0.02.

```python
from IPython.html.widgets import interact, interactive
from scipy.interpolate import interp1d 
```

```python
def Euler(dx):
    xmax = 2.
    y0 = 0.
    nsteps = int(xmax/dx) 
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    y[0] = y0
    
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + (x[i] + 2*y[i])*dx
    
    return x,y

xe,ye = Euler(0.25)
x2e,y2e = Euler(0.02)
```

```python
Interp = interp1d(xe,ye)
newX = 1.0
newfunc = Interp(newX)
plt.plot(xe,ye,newX,newfunc,"r+",markersize=20,markeredgewidth=3,linewidth=2.0)
plt.xlabel("Distance(m)",fontsize=20)
plt.ylabel("y(x)",fontsize=20)
plt.xlim(0.,2.)
plt.ylim(-10.,20.)
plt.grid()
plt.show()

print "Step Size 0.25: y(1) =",newfunc
```

```python
Interp = interp1d(x2e,y2e)
newX = 1.0
newfunc = Interp(newX)
plt.plot(x2e,y2e,newX,newfunc,"r+",markersize=20,markeredgewidth=3,linewidth=2.0)
plt.xlabel("Distance(m)",fontsize=20)
plt.ylabel("y(x)",fontsize=20)
plt.xlim(0.,2.)
plt.ylim(-10.,20.)
plt.grid()
plt.show()

print "Step Size 0.02: y(1) =",newfunc
```

We can use numerical solvers and widgets to explore the dynamics of a projectile that is subject to a drag force proportional to the square of its velocity - "quadratic drag".
The force on the projectile is given by
∑F⃗ =mg⃗ −cv2v^=mg⃗ −cvv⃗ =mg⃗ −cv2x+v2y‾‾‾‾‾‾‾√v⃗ =ma⃗ 
Where the second term is the drag force, in the direction opposite that of the velocity vector. Re-writing these in terms of the first-order differential equations for vx and vy gives
mv˙x=−cv2x+v2y‾‾‾‾‾‾‾√vx
mv˙y=mg−cv2x+v2y‾‾‾‾‾‾‾√vy
taking the positive y direction to be down. Notice that these are coupled equations. There are both vx and vy terms in each one. The drag coefficient c depends on the shape and density of the projectile. We can re-write this system of equations in terms of the terminal velocity - the point at which the gravitational force balances the drag force. This happens when mg=cv2. Let's denote the terminal velocity as vter=mg/c‾‾‾‾‾√ and re-write our equations in terms of vter.
v˙x=−gvxv2terv2x+v2y‾‾‾‾‾‾‾√
v˙y=−g(1+vyv2terv2x+v2y‾‾‾‾‾‾‾√)
(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far the projectile can travel in the x direction before landing.

```python
def deriv(y,t):
    a = 9.81
    vter = 44
    vx = y[0]
    vy = y[2]
    m = 1
    
    y1 = y[3]
    x1 = y[1]
    
    dvxdt = (-a/m) * (vx/vter**2) * sqrt(vx**2 + vy**2)
    dvydt = (-a/m) * (1 + vy/vter**2 * sqrt(vx**2 + vy**2))
    
    return np.array([dvxdt, vx, dvydt, vy])
```

```python
vx0 = 30*np.cos(50./180.*np.pi)
vy0 = 30*np.sin(50./180.*np.pi)
x0 = 0
y0 = 0
t = np.arange(0., 200.+0.01, 0.01)

result = odeint(deriv, np.array([vx0,x0,vy0,y0]), t)
vx = result[:,0]
vy = result[:,2]

plt.plot(t, vx, label="x-velocity")
plt.plot(t, vy, label="y-velocity")
plt.xlabel("Time (seconds)",fontsize=15)
plt.ylabel("Velocity (m/s)",fontsize=15)
plt.legend()
plt.show()
```
